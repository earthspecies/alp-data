```python
def get_taxonomy(sci_name: str) -> dict[str, str]:
    query = f"""SELECT TaxonID, Kingdom, Phylum, "Class", "Order", Family, Genus
    FROM gbif WHERE ScientificName = '{sci_name}';"""
    out = conn.sql(query).fetchall()
    if len(out) == 0:
        return None
    out = out[0]
    return {"gbifID": out[0],
            "kingdom": out[1],
            "phylum": out[2],
            "class": out[3],
            "order": out[4],
            "family": out[5],
            "genus": out[6],
           }
```

# Loop over barkley canyon dataset and add taxonomy

```python
from tqdm.notebook import tqdm
```

```python
taxonomy = {"gbifID": [],
            "kingdom": [],
            "phylum": [],
            "class": [],
            "order": [],
            "family": [],
            "genus": [],
            "species_common": [],
            "species_scientific": [],
            "taxonomic_name": []}

for i, row in tqdm(orig.iterrows(), total=len(orig)):
    detected_species = row["Species"].split("|")

    gbifID = []
    kingdom = []
    phylum = []
    _class = []
    order = []
    family = []
    genus = []
    species_common = []
    species_scientific = []
    taxonomic_name = []
    
    for d in detected_species:
        
        if d in HIGH_LEVEL_TAXA:
            species_sci = "Unknown"
            species_comm = "Unknown"
            
            species_common.append(species_comm)
            species_scientific.append(species_sci)
            continue
            
        else:
            species_sci = species_map_scientific.get(d, "Unknown")
            species_comm = species_map_common.get(d, "Unknown")
        
            species_common.append(species_comm)
            species_scientific.append(species_sci)
        
        tax = get_taxonomy(species_sci)
        if tax is None:
            continue
        taxname = tax["phylum"] + " "  + tax["class"] + " " + tax["order"] + " " + tax["family"] + " " + tax["genus"]
        
        gbifID.append(tax["gbifID"])
        kingdom.append(tax["kingdom"])
        phylum.append(tax["phylum"])
        _class.append(tax["class"])
        order.append(tax["order"])
        family.append(tax["family"])
        genus.append(tax["genus"])
        taxonomic_name.append(taxname)

    taxonomy["gbifID"].append(gbifID or pd.NA)
    taxonomy["kingdom"].append(",".join(kingdom) or pd.NA)
    taxonomy["phylum"].append(",".join(phylum) or pd.NA)
    taxonomy["class"].append(",".join(_class) or pd.NA)
    taxonomy["order"].append(",".join(order) or pd.NA)
    taxonomy["family"].append(",".join(family) or pd.NA)
    taxonomy["genus"].append(",".join(genus) or pd.NA)
    taxonomy["species_common"].append(",".join(species_common))
    taxonomy["species_scientific"].append(",".join(species_scientific))
    taxonomy["taxonomic_name"].append(",".join(taxonomic_name) or pd.NA)
```

## Cast the left and right times

```python
df["Left time (sec)"] = df["Left time (sec)"].astype("float32")
df["Right time (sec)"] = df["Right time (sec)"].astype("float32")
```

```python
start_times = []
end_times = []
drop_indices = []
for i, row in df.iterrows():
    try:
        l = row["Left time (sec)"]
        l = l.replace("s", "").replace(";",".").replace("-", "")
        l = float(l)
        r = row["Right time (sec)"]
        r = r.replace("s","").replace(";",".").replace("-", "")
        r = float(r)
        if not (r > l):
            drop_indices.append(i)
        else:
            start_times.append(l)
            end_times.append(r)
    except Exception as e:
        print(e)
        break
```
