```python
import pandas as pd
```

```python
traindf = pd.read_csv("/home/gagan/open_audio_datasets/macaques/annotations.csv")
```

# Make separate splits

```python
valdf = traindf[traindf["split"] == "valid"]
valdf.shape
```

```python
traindf = traindf[traindf["split"] == "train"]
traindf.shape
```

# Add sex identity column

```python
sex_mapping = {
    "TH": "female",
    "TW": "female",
    "SN": "female",
    "IO": "female",
    "MU": "male",
    "QU": "male",
    "BE": "male",
    "AL": "male"
}
```

```python
traindf["sex"] = traindf["class"].apply(lambda x: sex_mapping[x])
valdf["sex"] = valdf["class"].apply(lambda x: sex_mapping[x])
```

```python
traindf.shape
```

# Drop split column

```python
traindf.drop("split", axis=1, inplace=True)
valdf.drop("split", axis=1, inplace=True)
```

# Rename 'filename' to 'file_name'


```python
traindf.rename(columns={"filename": "file_name"}, inplace=True)
valdf.rename(columns={"filename": "file_name"}, inplace=True)
```

```python
traindf.columns
```

# Add local_path

```python
traindf["local_path"] = traindf["file_name"].apply(lambda x: "audio/" + x)
valdf["local_path"] = valdf["file_name"].apply(lambda x: "audio/" + x)
```

# Rename 'class' to individual_id so that it doesn't clash with taxonomy columns

```python
traindf.rename(columns={"class": "individual_id"}, inplace=True)
valdf.rename(columns={"class": "individual_id"}, inplace=True)
```

```python
traindf.columns
```

# Add taxonomy

```python
species_scientific = "Macaca mulatta"
species_common = "Rhesus macaque"
kindom = "Animalia"
phylum = "Chordata"
class_ = "Mammalia"
order = "Primates"
family = "Cercopithecidae"
genus = "Macaca"
taxonomic_name = "Chordata Mammalia Primates Cercopithecidae Macaca mulatta"
```

```python
L = len(traindf)
traindf_tax = {"gbifID": [2436604] * L,
               "species_scientific": [species_scientific] * L,
               "species_common": [species_common] * L,
               "kingdom": [kindom] * L,
               "phylum": [phylum] * L,
               "class": [class_] * L,
               "order": [order] * L,
               "family": [family] * L,
               "genus": [genus] * L,
               "taxonomic_name": [taxonomic_name] * L,
              }
traindf_tax = pd.DataFrame(traindf_tax)
```

```python
traindf_tax.shape
```

```python
L = len(valdf)
valdf_tax = {"gbifID": [2436604] * L,
               "species_scientific": [species_scientific] * L,
               "species_common": [species_common] * L,
               "kingdom": [kindom] * L,
               "phylum": [phylum] * L,
               "class": [class_] * L,
               "order": [order] * L,
               "family": [family] * L,
               "genus": [genus] * L,
               "taxonomic_name": [taxonomic_name] * L,
              }
valdf_tax = pd.DataFrame(valdf_tax)
```

## concatenate

```python
train_df = pd.concat([traindf.reset_index(drop=True), traindf_tax], axis=1)
```

```python
train_df.shape
```

```python
train_df.head(5)
```

```python
val_df = pd.concat([valdf.reset_index(drop=True), valdf_tax], axis=1)
val_df.columns
```

```python
val_df.head(4)
```

```python
val_df.shape
```

# Save 

```python
train_df.to_csv("/home/gagan/open_audio_datasets/macaques/train.csv", index=False)
val_df.to_csv("//home/gagan/open_audio_datasets/macaques/val.csv", index=False)
```
