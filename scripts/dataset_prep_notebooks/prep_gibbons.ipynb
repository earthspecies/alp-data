# Create file_name, local_path and selection_table dataframe

```python
data = {
    "file_name": [t + ".wav" for t in table_names],
    "selection_table": [],
    "local_path": ["audio/" + s + ".wav" for s in table_names]
}

for s in selection_tables:
    with open(s, "r") as f:
        data["selection_table"].append(f.read())
```

```python
data = pd.DataFrame(data)
data.head(3)
```

# Taxonomy

```python
species_scientific = "Hylobates muelleri"
species_common = "Mueller's Gibbon"
kindom = "Animalia"
phylum = "Chordata"
class_ = "Mammalia"
order = "Primates"
family = "Hylobatidae"
genus = "Hylobates"
taxonomic_name = "Chordata Mammalia Primates Hylobatidae Hylobates mulleri"
```

```python
L = len(data)
data = pd.concat([data, pd.DataFrame({
    "kingdom": [kindom] * L ,
    "phylum": [phylum] * L, 
    "class": [class_] * L,
    "order": [order] * L,
    "family": [family] * L,
    "genus": [genus] * L,
    "species_scientific": [species_scientific] * L,
    "species_common": [species_common] * L,
    "taxonomic_name": [taxonomic_name] * L,
})], axis=1)
```
