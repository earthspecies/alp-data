# Fix BEANS

What we want is to introduce a new column called `labels_as_list` that maps the label column as a list of str (or int or Any) rather than a comma separated list of strings.

```python
beans = Beans(split="train")
```

```python
splits = Beans.info.split_paths
```

```python
splits
```

```python
beans._data.head(5)
```

```python
def to_list(x):
    if not x or pd.isnull(x):
        return []
    if isinstance(x, str):
        label = x.split(",")
        return [x.strip() for x in label]

    if isinstance(x, int):
        return [x]

    raise TypeError("Unrecognized type")
```

```python
beans._data.loc[:, "label"] = beans._data["label"].apply(to_list)
```

### Run to_fly 

```python
data_frames = {}
```

```python
for k, v in splits.items():

    if k in data_frames:
        continue
        
    df = Beans(split=k)._data
    if "label" in df:
        df["labels_as_list"] = df.label.apply(to_list)
    elif "answer" in df:
        df["labels_as_list"] = df.answer.apply(to_list)
        
    data_frames[k] = df

    print(f"Done with split {k}")
```

## Save

```python
for k, df in data_frames.items():

    print(f"Saving split {k}")
    # get split address
    path = anypath(splits[k])
    
    if path.suffix == ".csv":
        df.to_csv(str(path), index=False)
    elif path.suffix == ".jsonl":
        df.to_json(str(path), lines=True, orient="records")

    print(f"Saved split {k}")
    
```

```python
birdsetdata_frames = {}
```

```python
for k, v in birdset_splits.items():
    
    print(f"Working on split {k}")
    if k in birdsetdata_frames:
        continue
        
    df = BirdSet(split=k)._data
    if "label" in df:
        df["labels_as_list"] = df.label.apply(to_list)
    elif "label" not in df and "text" in df:
        df["labels_as_list"] = df.text.apply(to_list)
        
    birdsetdata_frames[k] = df

    print(f"Done with split {k}")
```
