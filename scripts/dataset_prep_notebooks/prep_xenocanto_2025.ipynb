# Create the new metadata csv for the recent dump

```python
import os
from urllib.parse import urlparse
import numpy as np
from pathlib import Path
```

```python
downloaded_xc = os.listdir("/home/gagan/esp_projects/esp-data/scripts/xeno_canto_audio/")
len(downloaded_xc)
```

```python
audio = audio.reset_index(drop=True)
audio.head(5)
```

```python
audio = audio.drop(["level_0", "index"], axis=1)
audio.shape
```

```python
audio.head(3)
```

```python
total_audio_filenames = {os.path.basename(urlparse(row["identifier"]).path): i for i, row in audio.iterrows()}
```

```python
total_audio_filenames_stems = {Path(f).stem: i for f,i in total_audio_filenames.items()}
```

```python
list(total_audio_filenames_stems.keys())[:5]
```

```python
from tqdm.notebook import tqdm
import pdb
```

```python
new_xc_data = []
not_found = {}

for i, fname in tqdm(enumerate(downloaded_xc), total=len(downloaded_xc)):

    record = {}
    # look for the same file in audio
    if fname not in total_audio_filenames:
        # check again but this time with just the stem
        fstem = Path(fname).stem
        if fstem not in total_audio_filenames_stems:
            not_found[fname] = i
            continue
        else:
            # get the gbifID
            audio_rec = audio.iloc[total_audio_filenames_stems[fstem]]
            gbifID = audio_rec["gbifID"]
            record.update(audio_rec.loc[['gbifID','description', 'license', 'rightsHolder']].to_dict())
    else:
        # get the gbifID
        audio_rec = audio.iloc[total_audio_filenames[fname]]
        gbifID = audio_rec["gbifID"]
        record.update(audio_rec[['gbifID','description', 'license', 'rightsHolder']].to_dict())

    occurence_record = df[df["gbifID"] == gbifID]
    if len(occurence_record) > 1:
        raise Exception("What the fuck")
    occurence_record = occurence_record[['license', 'rightsHolder', 'sex', 'lifeStage', 'behavior',
                                         'kingdom', 'phylum', 'class', 'order', 'family', 'genus',
                                         'eventDate', 'eventTime', 'fieldNotes', 'species', 'locality']].iloc[0].to_dict()

    record.update(occurence_record)
    record["file_name"] = fname
    record["url"] = audio_rec["identifier"]
    new_xc_data.append(record)  
            
        
```

#### Find those rows of darwin multimedia where the URL matches the one's we've already downloaded in new_xc_data

```python
matching_inds = np.where(darwin_multi_sub["Identifier"].isin(urllist) == True)[0]
matching_inds.shape
```

#### Issue: do the rows of darwin_multi_sub match rows of darwin?

```python
coreid = darwin_multi_sub["CoreId"].iloc[matching_inds[0]]
darwin[darwin['id'] == coreid]
```

```python
latdec = [pd.NA for _ in range(len(new_xc_data))]
longdec = [pd.NA for _ in range(len(new_xc_data))]
vernacularname = [pd.NA for _ in range(len(new_xc_data))]
```

```python


for i in tqdm(range(len(matching_inds))):

    coreid = darwin_multi_sub["CoreId"].iloc[matching_inds[i]]
    dsub = darwin[darwin['id'] == coreid]
    idx = urllist.index(darwin_multi_sub["Identifier"].iloc[matching_inds[i]])
    
    assert len(dsub) == 1, "too many matches"
    assert dsub["rightsHolder"].iloc[0] == new_xc_data["rightsHolder"].iloc[idx], "Rights holder doesnt match"
    
    latdec[idx] = dsub["latitudeDecimal"].iloc[0]
    longdec[idx] = dsub["longitudeDecimal"].iloc[0]
    vernacularname[idx] = dsub["vernacularName"].iloc[0]
    
```

```python
sum([isinstance(x, pd._libs.missing.NAType) for x in latdec])
```

```python
new_xc_data_v2 = new_xc_data.assign(**{"latitudeDecimal": latdec, "longitudeDecimal": longdec, "vernacularName": vernacularname},)
```
