List registered datasets, instantiate one, inspect its `info`, and iterate.

## List all registered datasets

```python
from alp_data import list_registered_datasets, print_registered_datasets

names = list_registered_datasets()
print(f"{len(names)} registered datasets:")
for name in names:
    print(f"  - {name}")
```

## Instantiate a dataset

Lets look at the **BEANS** benchmark [https://arxiv.org/abs/2210.12300]


```python
from alp_data import Beans
```

We'll use the `dogs` subset of BEANS (small, quick to load).

```python
beans = Beans(split="dogs_test", sample_rate=16000)
print(f"Length: {len(beans)}")
```

### All available splits are under

```python
print(beans.available_splits)
```

### Lets have a quick look at the columns  

```python
beans.columns
```

## Inspect the `info` attribute

Every dataset carries a `DatasetInfo` describing name, owner, version, sources, license, splits.

```python
info = beans.info
print(f"name:        {info.name}")
print(f"owner:       {info.owner}")
print(f"version:     {info.version}")
print(f"license:     {info.license}")
print(f"sources:     {info.sources}")
print(f"description: {info.description}")
print(f"splits:      {list(info.split_paths)[:5]} ...")
```

## Random access

```python
%%time
sample = beans[0]
print("keys:", list(sample.keys()))
print("audio shape:", sample["audio"].shape)
```

## Iterate

```python
%%time
for i, sample in enumerate(beans):
    print(f"{i}: audio={sample['audio'].shape}, label={sample.get('label')}")
    if i >= 4:
        break
```

# Streaming mode

One of the advantages we offer is streaming data instead of having to download data first

```python
from alp_data import INaturalist
```

```python
inat = INaturalist(split="val", streaming=True, backend="pandas")  # takes 0 sec to instantiate
```

```python
print(len(inat))
```

```python
%%time
for i, sample in enumerate(inat):
    print(f"{i}: audio={sample['audio'].shape}, label={sample.get('species_scientific')}")
    if i >= 4:
        break
```
