## Build from yaml 

`dataset_from_config` reads a YAML file and returns `(dataset, transform_metadata)`.
The top-level key tells it what to build:
- `dataset:` → single dataset
- `concat:`  → `ConcatenatedDataset`
- `chain:`   → `ChainedDataset`

```python
from alp_data import dataset_from_config

dataset, transform_metadata = dataset_from_config("configs/beans_basic.yaml")
print(f"name:   {dataset.info.name}")
print(f"len:    {len(dataset)}")
print(f"meta:   {transform_metadata}")
```

```python
sample = dataset[0]
print("keys:", list(sample.keys()))
print("audio shape:", sample["audio"].shape)
```
